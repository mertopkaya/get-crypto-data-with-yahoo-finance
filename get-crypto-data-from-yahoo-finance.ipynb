```python
import pandas as pd
import requests

url = "https://finance.yahoo.com/cryptocurrencies/?offset=0&count=100"

r = requests.get(url)
df_list = pd.read_html(r.text) #parsesall the tables in webpages to a list
df = df_list[0]
df
```

```python
symbol_list = df["Symbol"].tolist()
symbol_list
```

```python
!pip install yfinance
import yfinance as yf
import pandas as pd


datetime_index = pd.date_range(start='2020-01-01', end='2021-03-02', freq='D')
    
portfolio = pd.DataFrame(index=datetime_index)

for symbol in symbol_list:
    symbol_data = (
        yf.download(symbol, start="2020-01-01", end="2021-03-02")['Close']
        .rename(symbol)
     
              
    )
    portfolio = pd.concat([portfolio, symbol_data], axis=1)


```

```python
portfolio.tail()
```

```python
#portfolio=portfolio.dropna(how='all')
portfolio = portfolio.dropna(axis='columns', how ='all')
portfolio
```
